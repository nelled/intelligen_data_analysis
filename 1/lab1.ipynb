### Pivot tables and graphics

We want to see how the instance are distributed between two categories: 'International plan' and 'Churn'. 

<b>Exercise 7:</b>
* Build the cross table between features using the method <i>crosstab</i>.
* Visualize the distribution for the feature 'Churn', depending on the value of the features 'International plan', 'Voice mail plan', and 'Customer service calls'.
<b>Hint:</b> Use commands plt.sublot and sns.subplots.

What do you see? What conclusions can be drawn? What feature (intuitively) can be more important for Churn prediction?

```python
pd.crosstab(df['International plan'], df['Churn'], margins=True)
# The majority of people who are unloyal do not have an international plan
# seems logical, youre getting an international plan when you are faithful to your provider
```

```python
f, ax = plt.subplots(nrows=3)
sns.countplot('International plan', hue='Churn', data=df,ax=ax[0])
sns.countplot('Voice mail plan', hue='Churn', data=df, ax=ax[1])
sns.countplot('Customer service calls', hue='Churn', data=df, ax=ax[2])

# Intuitively, the number of service calls schould be a good predictor. Lots of calls = lots of bad feelings
# about the product = change of provider.

# Other than that the majority of loyal clients does not have an international plan. This probably is caused
# by the fact that the overall number of subscribers to an int. plan is relatively low.

# People with no voice mail plan are also more likely to belong to the churn group. The same argument applies 
# the purchase of additional subscriptions usually means a certain degree of commitment

# The number of customer service calls seems to have some influence, almost half of the people with more than 3 calls
# belong to churn.
```
